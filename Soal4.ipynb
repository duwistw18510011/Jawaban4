```python
# a. Umur : merupakan umur dari ibu hamil
# b. Kelahiran ke- : merupakan kelahiran anak ke-... dari ibu tersebut
# c. Waktu Kelahiran : 0 = sesuai dengan hari perkiraan lahir(HPL), 1 = prematur, 2 = melebihi hari perkiraan lahir
# d. Tekanan_darah : 0 = untuk darah rendah, 1 = normal, 2 = tinggi
# e. Kelainan Jantung : merupakan kelainan jantung untuk anak yang telah lahir 0 = normal, 1 = kelainan
# f. Caesarian : 1 = Ya, 0 = Normal
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
import pickle
```

```python
data = pd.read_excel('D:/Tugas/KULIAH/SMT6/DataMining/UAS/dataset_UAS-main/dataset_UAS-main/caesarian.xlsx')
```

```python
df = pd.DataFrame(data)
df
```

```python
dis = []
for i in range(80):
    dis.append(math.sqrt((float(df.iloc[i]['Age'])-30)**2+
                        (float(df.iloc[i]['Delivery Number'])-1)**2+
                        (float(df.iloc[i]['Delivery Time'])-0)**2+
                        (float(df.iloc[i]['Blood of Pressure'])-1)**2))
```

```python
df['dis'] = dis
df
```

```python
df.sort_values('dis')
```

```python
x = df.sort_values('dis').head(5)
x
```

```python
y = x["Caesarian"]
y
```

```python
np.mean(y)
```
